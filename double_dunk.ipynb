```python
from stable_baselines3 import DQN
```

```python
train_env = gym.make("ALE/DoubleDunk-v5",mode=2)
model = DQN('CnnPolicy', train_env, verbose=0, learning_starts=1000, buffer_size=50000)
model.learn(total_timesteps=500000)
train_env.close()
```

```python
watch_env = gym.make("ALE/DoubleDunk-v5", render_mode="human",mode=2)
obs, info = watch_env.reset()
score = 0
episode = 1

for _ in range(1000):
    action, _ = model.predict(obs)
    obs, reward, terminated, truncated, info = watch_env.step(action)
    score += reward
    if terminated or truncated:
        print(f'Episode:{episode} Score:{score}')
        score = 0
        episode += 1
        obs, info = watch_env.reset()
```
